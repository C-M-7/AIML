```python
import pandas as pd
```

```python
data = pd.read_csv("titanic_train.csv")
data.head()
```

# Cleaning the Data

### As name, ticket, cabin, PassengerId wouldnt help us so dropping them but first making a copy of the data

```python
data_copy = data.copy()
data_copy.drop(["Name", "Ticket", "PassengerId", "Cabin"], axis=1, inplace=True)
data_copy.head()
```

```python
data_copy.shape
```

```python
data_copy.describe()
```

## Now for NaN Values we can clearly see that count of Age is less than 891

```python
data_copy["Age"].fillna(data["Age"].median(), inplace=True)
data_copy.describe()
```

## Cols contaning string arent in describe section so filling them independently

```python
data_copy["Embarked"].fillna("U", inplace=True)
data_copy.head()
```

# As we know we cant process further with strings so converting them using mapping
### Its basically denoting 0 for male and 1 for female and same going 1,2,3,4 for C,Q,S,U like that

```python
from sklearn import preprocessing as pp
```

```python
genders = {"male" : 0, "female" : 1}
data1 = [data_copy]

embarked = {"U" : 0, "S" : 1, "Q" : 2, "C" : 3}


for datasets in data1 :
    datasets['Sex'] = datasets['Sex'].map(genders)
    datasets['Embarked'] = datasets['Embarked'].map(embarked)

data_copy
```

# Now our training data is cleaned and now we can go for model implementation

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
# Getting x_train and y_train
x_train = data_copy.iloc[:,1:]
y_train = data_copy.iloc[:,0]
```

```python
x_train
```

```python
y_train
```

```python
model.fit(x_train,y_train)
```

# We have to clean the test data too so going for that now

```python
data_test = pd.read_csv("titanic_test.csv")
data_test_copy = data_test.copy()
data_test_copy.head()
```

```python
y_res = data_test_copy["PassengerId"]
y_res
```

```python
data_test_copy.drop(["Name", "Ticket", "PassengerId", "Cabin"], axis=1, inplace=True)
data_test_copy.head()
```

```python
data_test_copy.describe()
```

```python
data_test_copy['Age'].fillna(data_test_copy['Age'].median(), inplace=True)
data_test_copy['Fare'].fillna(data_test_copy['Fare'].mean(), inplace=True)
data_test_copy.describe()
```

```python
data_test_copy['Embarked'].fillna("U", inplace=True)
data_test_copy.head()
```

```python
# mapping the values for the test data
genders_test = {"male" : 0, "female" : 1}
embarked_test = {"U" : 0, "S" : 1, "Q" : 2, "C" : 3}
data11 = [data_test_copy]
for i in data11:
    i['Sex'] = i['Sex'].map(genders_test)
    i['Embarked'] = i['Embarked'].map(embarked_test)

data_test_copy.head()
```

```python
y_pred = model.predict(data_test_copy)
y_pred
```

```python
df = pd.DataFrame({"PassengerId" : y_res.values,
                   "Survived" : y_pred})
df
```
