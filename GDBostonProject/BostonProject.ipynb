```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("train.csv")
print(train_data.shape)
train_data
```

```python
# in Pandas we have to use .iloc for retrieving any specific row or col from dataset
x_train = train_data.iloc[:,0:13]
x_train
```

```python
y_train = train_data.iloc[:,13]
y_train
```

# As we have retrieved the train data in x and y terms so now we can implement our model

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
# Getting the inbuilt function of GD
```

```python
model.fit(x_train,y_train)
```

# Now getting the test data

```python
x_test = np.loadtxt("test.csv",delimiter=",")
print(x_test.shape)
x_test
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
np.savetxt("y_predict.csv", y_predict)
```
